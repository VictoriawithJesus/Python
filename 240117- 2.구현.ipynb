```python
상하좌우 이동 문제

여행가 A는 N * N 크기의 정사각형 공간 위에 서있다.
이 공간은 1 * 1 크기의 정사각형으로 나눠져있다. 
가장 왼쪽 위 좌표는 (1, 1)이며, 가장 오른쪽 아래 좌표는 (N, N)에 해당한다.

여행가A는 상, 하, 좌, 우 방향으로 이동할 수 있다. 
시작좌표는 항상 (1, 1)이다. 

계획서에는 하나의 줄에 띄어쓰기를 기준으로 하여
L,R,U,D 중 하나의 문자가 반복적으로 적혀져있다.
각 문자의 의미는 다음과 같다.

L: 왼쪽으로 한 칸이동
R: 오른쪽으로 한 칸 이동
U: 위쪽으로 한 칸 이동
D: 아래로 한 칸 이동
    
* 여행가 A가 N * N 크기의 정사각형 공간을 벗어나는 움직임은 무시된다.
예를 들어 (1,1) 위치에서 L 혹은 U를 만나면 무시된다. 

```

```python
[(1,1), (1,2), (1,3), (1,4), (1,5),
 (2,1), (2,2), (2,3), (2,4), (2,5),
 (3,1), (3,2), (3,3), (3,4), (3,5),
 (4,1), (4,2), (4,3), (4,4), (4,4),
 (5,1), (5,2), (5,3), (5,4), (5,5)]

이런식으로 좌표가 되어있음
```

```python
여행가 A에게 R->R->R->U->D->D
라고 되어있으면 
(3,4)에 도착하게됨 (4번째 U에서 점프 못하므로 못 벗어남)
결론은 (3,4)라고 적으면 됨.
```

```python
조건
1)번째 줄에 공간의 크기를 나타내는 N이 주어진다.
(1 <= N <= 100)

2)번째 줄에 여행가 A가 이동할 계획서 내용이 주어진다.
(1 <=이동횟수 <=100)

<출력 조건>
첫째 줄에 여행가 A가 도착할 지점의 좌표(X, Y)를 공백으로 구분하여 출력한다.
```

```python
#우선 N을 입력받기
n = int(input())

#초기 값 (1,1 설정)
x, y = 1, 1

#계획서를 입력 받기
plans = input().split() 
#띄어쓰기를 기준으로 L, R, U, D이런 알파벳 애들 입력받음. 


#l,R, U, D에 따른 이동 방향
move_types = ['L','R','U','D']

dx = [0, 0, -1,1]
dy = [-1,1, 0, 0]
#x와 y끼리 나올 수 있는 모든 조합을 만듦 (인덱스대로 쌍임.
#근데 왜 둘다 1은 없냐......아 몰라...

#이동계획을 하나씩 확인한다
for plan in plans:
    #(plan을 입력받음, L,R,U,D이런애들을 하나씩 넣음.)
    #이동 후 좌표를 구한다. 
    for i in range(len(move_types)):
        #웬지 모르나 move_types 갯수만큼 i를 돌리라는데? 근데 move_type은 겨우 4개가 끝인데...?
        #근데 그냥 교재 따라가보자. 
        if plan == move_types[i]:
            #if 문이 나왔네, move_types[i] 0부터 3까지 4번을 돌리겠징;;
            #각각 해당하는 애들 (L,R,U,D)에 해당하는 애들을 할당하라는 소리같은데...
            nx = x + dx[i]
            ny = y + dy[i]

    
    #예외 : 공간을 벗어나면 무시한다.
    if nx <1 or ny <1 or nx >n or ny >n:
        continue
    
    #앞에서 입력 받은 것 대로 이동해라
    x, y = nx, ny
#출력해라 
print(x, y)
```

```python
#우선 N을 입력받기
n = int(input())

#초기 값 (1,1 설정)
x, y = 1, 1

#계획서를 입력 받기
plans = input().split() 

move_types = ['L','R','U','D']

dx = [0, 0, -1,1]
dy = [-1,1, 0, 0]


for plan in plans:

    for i in range(len(move_types)):
     #move_types은 4개 들어있음 i는 0,1,2,3 순으로 들어감.  
        if plan == move_types[i]:

            nx = x + dx[i]
            ny = y + dy[i]


    if nx <1 or ny <1 or nx >n or ny >n:
        continue
    

    x, y = nx, ny

print(x, y)
```
